```python
import numpy as np
import pickle
from keras.preprocessing import image
import warnings
warnings.filterwarnings('ignore')
```

```python
model = open('training.pkl','rb')
classifier = pickle.load(model)
test = pickle.load(model)
```

```python
#Prediction of image
%matplotlib inline
import tensorflow
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np

# img1 = image.load_img('dataset/training_set/cats/cat.1.jpg', target_size = (64, 64))
# img1 = image.load_img('/home/admin1/Machine Learning/CNN/datasets/test_set/cats/cat.4061.jpg', target_size = (64, 64))
img1 = image.load_img('/home/admin1/Machine Learning/CNN/datasets/test_set/dogs/dog.4036.jpg', target_size = (64, 64))
# img1 = image.load_img('test/Cat/10.jpg', target_size=(64, 64))
img = image.img_to_array(img1)
img = img/255
# create a batch of size 1 [N,H,W,C]
img = np.expand_dims(img, axis=0)
prediction = classifier.predict(img, batch_size=None,steps=1) #gives all class prob.
if(prediction[:,:]>0.5):
   value ='Dog :%1.2f'%(prediction[0,0])
   plt.text(20, 62,value,color='red',fontsize=18,bbox=dict(facecolor='white',alpha=0.8))
else:
   value ='Cat :%1.2f'%(1.0-prediction[0,0])
   plt.text(20, 62,value,color='red',fontsize=18,bbox=dict(facecolor='white',alpha=0.8))

plt.imshow(img1)
plt.show()
```
