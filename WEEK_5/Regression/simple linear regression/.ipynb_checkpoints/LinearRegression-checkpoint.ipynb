```python
# data preprocessing on given dataset
import pandas as pd
import numpy as np

class LinearRegression:
    def __init__(self):
        self.learning_rate = 0.1
        self.iteration = 1000
        self.theta_0 = 0
        self.theat_1 = 0
        self.epoch = 10000

    def handle_missing_data(self):
#         replace missing data with mean
        print("Train dataset")
#         print("\nsum of NaN  values in dataset\n\n",self.dataset_train.isnull().sum())
        print("\nReplace missing data with mean")
        self.dataset_train.replace(np.NaN,self.dataset_train.mean(),inplace=True)
        print("\n",self.dataset_train)
        print("\n\n\n")
        print("\nTest dataset")
#         print("\nsum of NaN  values in dataset\n\n",self.dataset_train.isnull().sum())
        print("\nReplace missing data with mean")
        self.dataset_test.replace(np.NaN,self.dataset_test.mean(),inplace=True)
        print("\n",self.dataset_test)
       
    def feature_scaling(self):
#         standerdization of dataset
        for column in self.dataset_train.columns:
            self.dataset_train[column] = (self.dataset_train[column] - self.dataset_train[column].min()) / (self.dataset_train[column].max()-self.dataset_train[column].min())
        print("\n\nNormalization\n\n",self.dataset_train)
#         Normalization of  dataset
        for name in self.dataset_train.columns:
            self.dataset_train[name] = (self.dataset_train[name] - self.dataset_train[name].mean()) / np.nanstd(self.dataset_train[name])
        print("Standerdization\n\n", self.dataset_train)
        
    def gradient_descent(self,x_data,y_data):
        n = len(y_data)
        theta_0 = self.theta_0
        theta_1 = self.theat_1
        for i in  range(self.epoch):
            for data  in range(len(x_data)):
                hypo = theta_0+theta_1*x_data[data]
                h1 = hypo-y_data[data]
                h2 = (hypo-y_data[data])*x_data[data]
                temp1 = ((self.learning_rate*h1)/len(x_data))
                temp2 = ((self.learning_rate*h2)/len(x_data))
            gd_theta_0 = theta_0-temp1
            gd_theta_1 = theta_1-temp2
        return gd_theta_0,gd_theta_1
    
    def test(self, x_test_data, theta_0, theta_1):
        y_predict = [None]*len(x_test_data)
        for i in range(len(x_test_data)):
            y_predict[i] = theta_1*x_test_data[i] + theta_0

        return y_predict

    def accuracy(self, y_test_data, y_predicted):
        diff = 0
        for i in range(len(y_train_data)):
            diff += abs((y_predicted[i] - y_test_data[i])/y_train_data[i])

        data_accuracy = diff*100
        return data_accuracy        
        
        
def main():
    obj =LinearRegression()
    x_train_data = np.array(train_data["x"].values)
    y_train_data = np.array(train_data["y"].values)
    
    x_test_data = np.array(test_data["x"])
    y_test_data = np.array(test_data["y"])

    theta_0, theta_1= obj.gradient_descent(x_train_data,y_train_data)
    print(theta_0,theta_1)
    
    y_prediction =obj.test(x_test_data,theta_0,theta_1)
    accuracy = obj.accuracy(y_test_data,y_prediction)
    print("Accuracy:",accuracy)
if __name__ == '__main__':
    main()
```
